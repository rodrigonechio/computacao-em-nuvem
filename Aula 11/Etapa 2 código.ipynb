```python
from googletrans import Translator

translator = Translator()
text_to_translate = "Hello, world! How are you today?"

languages = ['es', 'fr', 'de', 'it', 'ja'] # Spanish, French, German, Italian, Japanese

print("Original text:", text_to_translate)

for lang in languages:
    translated_text = translator.translate(text_to_translate, src='en', dest=lang)
    print(f"Translation ({lang}):", translated_text.text)


```
